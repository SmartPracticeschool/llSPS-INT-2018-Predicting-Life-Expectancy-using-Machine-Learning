```python
# seperating dependent and independent featues 

X = df.drop('Life expectancy ', axis = 1)
y = df['Life expectancy ']
X.head()
print(y.head())
print(X.head())
```

```python
# count of categoricl features in country column

categ_feature=len(df['Country'].unique())
print(categ_feature)
```

```python
cntdum=pd.get_dummies(X['Country'])
stadum=pd.get_dummies(X['Status'])
X.drop('Country',inplace=True,axis=1)
```

```python
X.drop('Status',inplace=True,axis=1)
X=pd.concat([X,cntdum,stadum],axis=1)
```

```python
#train-test split


from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
#fitting model

from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
#predicting values using trained model

y_pred=reg.predict(X_test)
```

```python
#side by side comparison of first few predicted and true value

print("Actual Value    Predicted value")

k=0
for i in Y_test:
    for j in range(len(y_pred)):    
        print(i,"          ",y_pred[k])
        k+=1
        break
    if k==15:
        break
```

```python
from sklearn import metrics
```

```python
# rmse error 

print("RMSE ERROR: ",np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))  
```

```python
#accuracy of model

res=reg.score(X_test,Y_test)
print("Accuarcy : ",res*100)
```
